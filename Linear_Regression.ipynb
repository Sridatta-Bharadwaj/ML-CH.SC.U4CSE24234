# Linear Regression implementation

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
import kagglehub

path = kagglehub.dataset_download("ravitejakotharu/salary-datacsv")

print("Path to dataset files:", path)
```

```python
#reading the csv
df = pd.read_csv(f'{path}/salary_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# checking if there are any missing values
df.isnull().sum()
```

```python
#defining the dependent and the independent columns
x = df.iloc[:, :1].values
y = df.iloc[:, 1:2].values
```

```python
# splitting the dataset into training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# fitting the dataset to the model and predicting values
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
#plotting the dataset and the linear regression line (for training set)
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience (Training Set)')
plt.show()
```

```python
# plotting the dataset and the linear regression line (for testing set)
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience (Testing Set)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
data = pd.read_csv('/Weight_prediction.csv')
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Gender'] = data['Gender'].replace('Ma le', 'Male')
data['Gender'] = data['Gender'].str.strip()

data_encoded = pd.get_dummies(data, columns=['Gender'], drop_first=True, dtype=int)

x = data_encoded[['Height', 'Age', 'Gender_Male']].values
y = data_encoded[['Weight']].values
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

```python
plt.scatter(x_train[:, 0], y_train, color='blue')

height_min, height_max = x_train[:, 0].min(), x_train[:, 0].max()
height_range = np.linspace(height_min, height_max, 100).reshape(-1, 1)

mean_age = x_train[:, 1].mean()
mean_gender_male = x_train[:, 2].mean()

dummy_x_for_plot = np.hstack([height_range,
                              np.full(height_range.shape, mean_age),
                              np.full(height_range.shape, mean_gender_male)])

predicted_weights_for_plot = model.predict(dummy_x_for_plot)

plt.plot(height_range, predicted_weights_for_plot, color='red')

plt.title('Weight vs Height (Training Set)')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
plt.scatter(x_test[:, 0], y_test, color='blue')

height_min_test, height_max_test = x_test[:, 0].min(), x_test[:, 0].max()
height_range_test = np.linspace(height_min_test, height_max_test, 100).reshape(-1, 1)

dummy_x_for_plot_test = np.hstack([height_range_test,
                                   np.full(height_range_test.shape, mean_age),
                                   np.full(height_range_test.shape, mean_gender_male)])

predicted_weights_for_plot_test = model.predict(dummy_x_for_plot_test)

plt.plot(height_range_test, predicted_weights_for_plot_test, color='red')

plt.title('Weight vs Height (Testing Set)')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
